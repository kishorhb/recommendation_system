```python
book_user= book.pivot_table(index='user_id',
                                 columns='book_tittle',
                                 values='book_rating').reset_index(drop=True)
```

```python
book_user
```

```python
book_user.index = book.user_id.unique()
```

```python
book_user
```

```python
#Impute those NaNs with 0 values
book_user.fillna(0, inplace=True)
```

```python
book_user
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( book_user.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book.user_id.unique()
user_sim_df.columns = book.user_id.unique()
```

```python
user_sim_df.iloc[0:10, 0:10]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:10, 0:10]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:10]
```
